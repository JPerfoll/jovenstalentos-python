```python
import pandas as pd
```

```python
globalstore = pd.read_csv('globalstore.csv')
```

```python
mercados = pd.DataFrame()
primeiro = 0
for mercado in ['LATAM', 'US', 'Canada']:
    if primeiro == 0:
        mercados = globalstore[globalstore['Market'] == mercado]
    else:
        mercados = pd.concat(
            [
                mercados,
                globalstore[globalstore['Market'] == mercado]
            ]
        )

mercados
```
